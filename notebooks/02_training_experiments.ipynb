# G-NeSAI — Training Experiments

This notebook loads training artifacts from `outputs/`, visualizes metrics, and reproduces key figures for publication.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path

OUT = Path('..') / 'outputs'
cm_path = OUT / 'confusion_matrix_eval.png'
rpt_path = OUT / 'classification_report_eval.json'
ablation_csv = OUT / 'ablation_summary.csv'

print('Artifacts:')
print(cm_path.resolve())
print(rpt_path.resolve())
print(ablation_csv.resolve())
```

```python
from IPython.display import Image, display
if cm_path.exists():
    display(Image(filename=str(cm_path)))
else:
    print('Confusion matrix image not found; run training/eval_and_plots.py')
```

```python
if rpt_path.exists():
    with open(rpt_path) as f:
        rpt = json.load(f)
    print('Classification report sample (per-class precision/recall/f1):')
    for k,v in rpt.items():
        if k.isdigit() or k in ['0','1','2'] or k.startswith('BPSK') or isinstance(v, dict):
            print(k, v if isinstance(v, dict) else '')
else:
    print('Report not found')
```

```python
import pandas as pd
if ablation_csv.exists():
    df = pd.read_csv(ablation_csv)
    display(df)
else:
    print('Ablation CSV not found')
```
